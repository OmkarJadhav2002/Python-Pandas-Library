```python
# API2
def team_vs_team(team1, team2):
    temp_df = ipl[(ipl["Team1"] == team1) & (ipl["Team2"] == team2) | (ipl["Team1"] == team2) & (ipl["Team2"] == team1)]
    
    total_matches = temp_df.shape[0]
    matches_won_team1 = temp_df["WinningTeam"].value_counts()[team1]
    matches_won_team2 = temp_df["WinningTeam"].value_counts()[team2]
    
    draw_macthes = total_matches - (matches_won_team1 + matches_won_team2)
    
    response = {
        "total_matches_played": total_matches,
        team1: matches_won_team1,
        team2: matches_won_team2,
        "draw matches": draw_macthes
    }
    
    return response
```

```python
team_vs_team("Royal Challengers Bangalore", "Chennai Super Kings")
```
