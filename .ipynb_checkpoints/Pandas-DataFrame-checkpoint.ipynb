```python
points_df = pd.DataFrame({
    "1st Point": [(-5, 2), (0, 7), (2, 3), (3, 4), (5, -2), (5, 6)],
    "2nd Point": [(2, 3), (1, -9), (5, 8), (2, 3), (1, 0), (0, 0)]
})
```

```python
points_df
```

```python
def euclidian(row):
    pt_a = row["1st Point"]
    pt_b = row["2nd Point"]
    
    return ((pt_a[0] - pt_b[0])**2 + (pt_a[1] + pt_b[1])**2)**0.5
```

```python
points_df["Euclidian_Distance"] = points_df.apply(euclidian, axis=1)
```

```python
points_df
```

### Groupby()
* study of groups
* always applicable to categorical features/columns 
* you make/ form groups in a data on the basis of column


**convert to groups -> get the column -> aggregate function**

```python
import numpy as np
import pandas as pd
import time
```

```python
films = pd.read_csv("imdb-top-1000.csv")
films.head(2)
```

```python
genres = films.groupby("Genre") 
```

```python
genres.max()
```

```python
# find the top 3 genres by total earning
# films.groupby("Genre").sum()["Gross"].sort_values(ascending=False).head(3)

# get column first and then apply the aggregate functions
# for bigger datasets (order of code matters)
start = time.time()
films.groupby("Genre").sum()["Gross"].sort_values(ascending=False).head(3)
print("Time: ", time.time()-start)
```

```python
import time
start = time.time()
films.groupby("Genre")["Gross"].sum().sort_values(ascending=False).head(3)
print("Time: ", time.time()-start)
```

```python
# find the genre with highest avg IMDB rating
films.groupby("Genre")["IMDB_Rating"].mean().sort_values(ascending=False).head(3)
```

```python
# find the director with most popularity
films.groupby("Director")["No_of_Votes"].sum().sort_values(ascending=False).head(1)
```

```python
# find highest rated movie of each genre
films.groupby("Genre")[["IMDB_Rating", "Series_Title"]].max().sort_values(ascending=False)
```

```python
start = time.time()
films.groupby("Star1")["Series_Title"].sum().sort_values(ascending=False)
print("Time required: ", time.time()-start)
```

```python
# find the number of movies done by each actor
# films["Star1"].value_counts()
films.groupby("Star1")["Series_Title"].count().sort_values(ascending=False).head(5)
```

#### split (apply) combine
* you can pass your own logic to solve a particular question using apply (you can pass function), (custom logic)

```python
genres.apply(min) # built in function
```

```python
# find number of movies starting with A for each group
# df = pd.DataFrame(columns=films.columns)
def find(group):
    # df = pd.DataFrame(columns=films.columns)
    # df = df.append(group[group["Series_Title"].str.startswith('A')])
    
    return group["Series_Title"].str.startswith('A').sum() 
```

```python
genres.apply(find)
```

```python
# find ranking of each movie in the group according to IMDB score

def rank(group):
    group["Genre_Rank"] = group["IMDB_Rating"].rank(ascending=False)
    return group
```

```python
genres.apply(rank)
```

```python
# find normalized IMDB rating group wise

def normal(group):
    group["Norm_Rating"] = (group["IMDB_Rating"] - group["IMDB_Rating"].min())/(group["IMDB_Rating"].max() - group["IMDB_Rating"].min())
    return group
```

```python
genres.apply(normal).sample(10)
```

```python
# Create a function that can return the highest score of any batsman

# temp_df = ipl[ipl["batsman"] == "V Kohli"]
# temp_df.groupby("match_id")["batsman_runs"].sum().sort_values(ascending=False).head(1).values[0]

def high_score(batsman):
    temp_df = ipl[ipl["batsman"] == batsman]
    return temp_df.groupby("match_id")["batsman_runs"].sum().sort_values(ascending=False).head(1).values[0]
```
