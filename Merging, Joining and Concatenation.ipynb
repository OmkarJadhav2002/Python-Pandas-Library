### Questions

```python
# Q.1 Find total revenue generated
# courses.head()
total_reg.merge(courses, how="inner", on="course_id")["price"].sum()
# courses.merge(m2, how="inner", on="course_id")
# m1["level_0"]
```

```python
# find month by month revenue
temp_df = pd.concat([m1, m2], keys=["Nov", "Dec"]).reset_index()
# temp_df
temp_df.merge(courses, how="inner", on="course_id").groupby("level_0")["price"].sum()
# courses.merge(m1, how="inner", on="course_id")["price"].sum()
```

```python
# Print the registration table
# cols = name, course, price
total_reg.merge(stud, on="student_id").merge(courses, on="course_id")[["name", "course_name", "price"]]
```

```python
# Plot the bar chart for revenue and course
total_reg.merge(courses, on="course_id").groupby("course_name")["price"].sum().plot(kind="bar", color="salmon");
```

```python
# find students who enrolled in both the months
# total_reg.merge(m1, on="student_id")
common_student = np.intersect1d(m1["student_id"], m2["student_id"])
```

```python
# stud[stud["student_id"].isin(common_student)]
courses
# stud
# total_reg
```

```python
# find course that got no enrollment
# courses["course_name"]
course = np.setdiff1d(courses["course_id"], total_reg["course_id"])
courses[courses["course_id"].isin(course)]
```

```python
# find sudents who did not enroll into any courses
not_enrolled =np.setdiff1d(stud["student_id"], total_reg["student_id"])
stud[stud["student_id"].isin(not_enrolled)].shape[0]
```

```python
# Print students name -> partner name for all enrolled students
# total_reg.merge(stud, on="student_id")[["name", "partner"]]
# total_reg.merge(stud, how="self", on="student_id")

# self join: joining table with itself
stud.merge(stud, how="inner", left_on="partner", right_on="student_id")[["name_x","name_y"]]
```

```python
# find top 3 students who did most number of enrollments
total_reg.merge(stud, on="student_id").groupby(["student_id", "name"])["name"].count().sort_values(ascending=False).head(3)
```

```python
# find top 3 students who spent most amount of money on courses
# total_reg.merge(courses, on="course_id").groupby(["course_id", "student_id"])["price"].sum().sort_values(ascending=False).head(3)
total_reg.merge(stud, on="student_id").merge(courses, on="course_id").groupby(["student_id", "name"])["price"].sum().sort_values(ascending=False).head(3)
```
